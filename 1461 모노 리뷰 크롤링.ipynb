```python
chromedriver_autoinstaller.install(cwd=True)
driver = webdriver.Chrome()   # driver : 크롬 브라우저

driver.get('https://brand.naver.com/drmartens/products/5316380050')
time.sleep(2)

# 리뷰 클릭
selector = "#content > div > div.Q1bBXdV7RJ > div.MV8TIZ8uR3 > ul > li:nth-child(2)"
driver.find_element(By.CSS_SELECTOR, selector).click()

# 1p ~ 10p 리뷰 크롤링
review_list = []

for i in range(0, 12):

    try:
        # 페이지 클릭 (1p = nth-child(2))
        selector = f"#REVIEW > div > div.JHZoCyHfg7 > div.HTT4L8U0CU > div > div > a:nth-child({i})"
        driver.find_element(By.CSS_SELECTOR, selector).click()
        time.sleep(2)

        # 리뷰 수집
        selector = ".KqJ8Qqw082"
        review = driver.find_elements(By.CSS_SELECTOR, selector)

        for j in review:
            review = j.text
            review_list.append(review)

        print(f"{i-1}p 수집 완료 / 누적 {len(review_list)}개")

    except:
        continue

print(len(review_list))
review_list
```
